```python
import pandas as pd
df = pd.read_csv("Table data.csv")
df
```

```python
df = df[df["Content"] != "Total"].copy()
```

```python
df.isnull().sum()
```

```python
df = df.dropna(subset = ["Video title", "Video publish time"])
```

```python
df
```

```python
df["Video publish time"] = pd.to_datetime(df["Video publish time"], errors = 'coerce')
```

```python
df
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_").str.replace("(", "").str.replace(")", "")
```

```python
df
```

```python
df = df.drop_duplicates()
```

```python
df
```

```python
df.dtypes
```

```python
df = df.sort_values(by = "views", ascending = False)
```

```python
df
```

```python
df = df.reset_index(drop = True)
```

```python
df
```

```python
df.info()
```

```python
df.head()
```

```python
df["year"] = df["video_publish_time"].dt.year
```

```python
df["month"] = df["video_publish_time"].dt.month
```

```python
df
```

```python
df["day_name"] = df["video_publish_time"].dt.day_name()
# which day video perfom best
```

```python
df
```

```python
df["views_per_subscribers"] = df["views"] / df["subscribers"]
```

```python
df
```

```python
df["engagement"] = df["watch_time_hours"] / df["views"]
```

```python
df
```

```python
import numpy as np

df["views_per_subscribers"] = df["views"] / df["subscribers"]
df["views_per_subscribers"] = df["views_per_subscribers"].replace([np.inf, -np.inf], 0)
```
